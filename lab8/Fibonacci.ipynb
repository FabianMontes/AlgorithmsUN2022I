recursive Fibonacci

```python
import itertools as tools
import matplotlib.pyplot as plt
from time import time
```

```python

def fiboRec(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fiboRec(n-1)+fiboRec(n-2) 

def fiboItera(n):
  a=0
  b=1
  for i in range(n):
    a=b
    b=a+b
  return a

def fiboLog(n):
  if n==0: return 0   
  v1=1
  v2=1
  v3=0  
  for rec in bin(n)[3:]:
      calc = (v2*v2) 
      v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3) 
      if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
  return v2;
```

```python
from time import process_time

def timeFib(n,t):
  t0,tf =0 ,0
  if(t=='r'):
    t0 = process_time()
    fiboRec(n)
    tf =process_time()
  else:
      if(t=='i'):
        t0 = process_time()
        fiboItera(n)
        tf =process_time()
      else:
        t0 = process_time()
        fiboLog(n)
        tf =process_time()
  
  return tf-t0
```

```python
#recursive Fibonacci time
Tim = []
numb = []
for i in range(50):
  numb.append(i)
  Tim.append(timeFib(i,'r'))
plt.title("Fibonacci Iterativo")
plt.plot(numb,Tim,label="Fibonacci Recursivo")
plt.plot(numb,numb,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```

```python
#Iterative Fibonacci lineartime
timeI = []
numbI = []
for i in range(1000):
  numbI.append(i)
  timeI.append(timeFib(i,'i'))
plt.title("Fibonacci Iterativo")
plt.plot(numbI,timeI,label="Fibonacci Iterativo")
plt.plot(numbI,numbI,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```

```python
#iterative Fibonacci logarithmic time
timeL = []
numbL = []
for i in range(1000):
  numbL.append(i)
  timeL.append(timeFib(i,'l'))
plt.title("Iterative Fibonacci logarithmic time")
plt.plot(numbL,timeL,label="Fibonacci Iterativo logaritmico")
plt.plot(numbL,numbL,label="Ejecucion lineal")
plt.ylabel("Tiempo")
plt.show()
```
